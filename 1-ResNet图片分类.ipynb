```python
from torchvision import models
```

```python
dir(models)
```

```python
resnet = models.resnet101(pretrained=True)
```

```python
resnet
```

```python
from torchvision import transforms
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
    mean = [0.485, 0.456, 0.406],
    std = [0.229, 0.224, 0.225]
    )
])
```

```python
from PIL import Image
```

```python
# img = Image.open('./data/1/dog1.jpg')
# img = Image.open('./data/1/dog2.jpg')
img = Image.open('./data/1/dog3.jpg')
```

```python
img
```

```python
img_t = preprocess(img)
```

```python
import torch
batch_t = torch.unsqueeze(img_t, 0)
```

```python
resnet.eval()
```

```python
out = resnet(batch_t)
```

```python
out
```

```python
with open('./data/1/imagenet_classes.txt') as f:
    lables = [line.strip() for line in f.readlines()]
```

```python
_, index = torch.max(out, 1)
```

```python
percentage = torch.nn.functional.softmax(out, dim = 1)[0] * 100
```

```python
lables[index[0]], percentage[index[0].item()]
```

```python
_, indices = torch.sort(out, descending = True)
[(lables[idx], percentage[idx].item()) for idx in indices[0,:5]]
```
